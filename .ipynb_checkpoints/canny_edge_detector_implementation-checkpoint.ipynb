## This is an implementation of the Canny Edget 
The test image is the cameraman image which comes as a part of OpenCV and is displayed in grayscale and smoothened using the gaussian smoothing operator.
This is divided into 2 subtasks
1. Creation of a gaussian kernel with specified parameters
    - The parameter **sigma (Ïƒ)** is the standard deviation of the gaussian filter.
    - The **size** parameter is used to determine the size of the kernel.
    - The **range_gaussian** parameter scales the kernel.
2. Reading the cameraman image and applying the kernel to it

```python
%matplotlib inline
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider, IntSlider, Checkbox, Layout, HBox, Label
from skimage import data
from scipy.ndimage import convolve  # this is a multidimensional convolution different from the one in scipy.signalss

# normalization of the cameraman image to values between 0 and 1
original_image = data.camera()/255
# global variables
smoothened_image = np.zeros((len(original_image), len(original_image[0])))
magnitude_edge_image, direction_edge_image = [], []
suppressed_image = np.zeros((len(original_image), len(original_image[0])))
angle_array = np.array([0, (np.pi)/4, (np.pi)/2, 3*(np.pi)/4])


# my guess is that the number of functions in the gaussian function below determine the number of sliders
def display_gaussian_kernel(sigma=5.0, size=7, range_gaussian=5):
    global smoothened_image
    X = np.linspace(-range_gaussian, range_gaussian, size)
    Y = np.linspace(-range_gaussian, range_gaussian, size)
    # generate a grid of X, Y coordinates with the size indicating the kernel size
    X, Y = np.meshgrid(X, Y)
    # mu = np.array([np.mean(X), np.mean(Y)])
    # we know that the mean will always be zero for any value of range_gaussian and size
    mu = np.array([0.0, 0.0])
    # A covariance matrix of the form, Î£=(Ïƒ^2*I) has variance = Ïƒ^2 for all variables
    covariance = np.diag(np.array([sigma, sigma])**2)
    # generate XY coordinates from the X and Y arrays
    XY = np.column_stack([X.flat, Y.flat])
    z = scipy.stats.multivariate_normal.pdf(XY, mean=mu, cov=covariance)
    gaussian_kernel = z.reshape(X.shape)
    # Plot the surface.
    fig1 = plt.figure()
    fig1.set_size_inches(12, 6)
    ax1 = fig1.add_subplot(131)
    ax1.set(title='Gaussian Kernel')
    ax1.imshow(gaussian_kernel)
    ax2 = fig1.add_subplot(132)
    ax2.set(title='Original Image')
    ax2.imshow(original_image, cmap=plt.cm.gray)
    ax2.axis('off')
    ax3 = fig1.add_subplot(133)
    ax3.set(title='Smoothened Image')
    smoothened_image = convolve(original_image, gaussian_kernel)
    # renormalizing by rescaling
    smoothened_image = smoothened_image/max(smoothened_image.flatten())
    ax3.imshow(smoothened_image, cmap=plt.cm.gray)
    ax3.axis('off')
    plt.show()
    
interact(display_gaussian_kernel,
         sigma=FloatSlider(min=1, max=10, step=0.05, value=5.0, continuous_update=False),
         size=IntSlider(min=3, max=61, step=2, value=7.0, continuous_update=False),
         range_gaussian=IntSlider(min=5, max=15, step=1, value=5.0, continuous_update=False))
```

Plots displayed:
1. The smoothened image
2. The edges that we extracted from that image using the sobel operator
3. The direction of the vector gradients in the image

```python
def sobel_operator():
    global magnitude_edge_image, direction_edge_image
    horizontal_kernel = np.array([[1.,  2,  1], [0,  0,  0], [-1, -2, -1]]) # Gy
    vertical_kernel = np.array([[-1.,  0,  1], [-2,  0,  2], [-1, 0, 1]])   # Gx
    horizontal_edge_image = convolve(smoothened_image, horizontal_kernel)
    vertical_edge_image = convolve(smoothened_image, vertical_kernel)
    magnitude_edge_image = np.sqrt(horizontal_edge_image**2 + vertical_edge_image**2)
    # renormalize by scaling
    magnitude_edge_image = magnitude_edge_image/max(magnitude_edge_image.flatten())
    # calculate the direction of edges and also round off angles to the nearest number
    with np.errstate(divide='ignore', invalid='ignore'):
        direction_edge_image = np.abs(np.arctan(np.true_divide(horizontal_edge_image, vertical_edge_image)))
        direction_edge_image = direction_edge_image.flatten()
        for index in range(len(direction_edge_image)):
            direction_edge_image[index] = angle_array[(np.abs(angle_array - direction_edge_image[index])).argmin()]
        direction_edge_image = direction_edge_image.reshape((len(smoothened_image), len(smoothened_image[0])))
       
    fig2 = plt.figure()
    fig2.set_size_inches(12, 12)
    ax1 = fig2.add_subplot(221)
    ax1.set(title='Smoothened Image')
    ax1.axis('off')
    ax1.imshow(original_image, cmap=plt.cm.gray)
    ax2 = fig2.add_subplot(222)
    ax2.set(title='Magnitude of Edges')
    ax2.imshow(magnitude_edge_image, cmap=plt.cm.gray)
    # using quiver to plot the edge direction
    X, Y = np.meshgrid(np.arange(0, len(original_image), 1), np.arange(0, len(original_image[0]), 1))
    # renormalizing U, V would be overkill
    U = magnitude_edge_image*np.cos(direction_edge_image)
    V = magnitude_edge_image*np.sin(direction_edge_image)
    ax3 = fig2.add_subplot(223)
    Q = ax3.quiver(X, Y, V, U)
    l, r, b, t = ax3.axis()
    # shorten the image so that it fits within the border
    ax3.axis([l, r-(r-len(original_image)), b, t-(t-len(original_image))])
    ax3.set(title='Direction of Edges')
    ax3.set_ylim(ax3.get_ylim()[::-1])
    plt.show()
    
interact(sobel_operator)
```

The non maximum suppression algorithm is applied on the previously obtained image (Magnitude of Edges).


```python
def non_max_suppression():
    global suppressed_image
    # four filters for the various directions
    sup_filter_zero_deg = np.array([[0, 0, 0], [-1, 2, -1], [0, 0, 0]])
    sup_filter_fortyfive_deg = np.array([[0, 0, -1], [0, 2, 0], [-1, 0, 0]])
    sup_filter_ninety_deg = np.array([[0, -1, 0], [0, 2, 0], [0, -1, 0]])
    sup_filter_onethirtyfive_deg = np.array([[-1, 0, 0], [0, 2, 0], [0, 0, -1]])
    sup_filters = [sup_filter_zero_deg, sup_filter_fortyfive_deg, sup_filter_ninety_deg, sup_filter_onethirtyfive_deg]
    
    # slow but easy to understand
    # if the direction of the gradient at a particular point is not the same as the filter its being convolved
    # with then the value at that point is set to 0
    for index, sup_filter in enumerate(sup_filters):
        temp = convolve(magnitude_edge_image, sup_filter)
        mag_copy = magnitude_edge_image.copy()
        for i in range(len(temp)):
            for j in range(len(temp[i])):
                if (direction_edge_image[i][j] != angle_array[index]) or (temp[i][j] <= 0):
                    temp[i][j] = 0             # suppression
                else:
                    temp[i][j] = 1
        suppressed_image += temp*mag_copy      # add the results of applying all four filters
    
    # normalize by rescaling
    suppressed_image = suppressed_image/max(suppressed_image.flatten())
    fig3 = plt.figure()
    fig3.set_size_inches(12, 6)
    ax1 = fig3.add_subplot(121)
    ax1.set(title='Magnitude of Edges')
    ax1.axis('off')
    ax1.imshow(magnitude_edge_image, cmap=plt.cm.gray)
    ax2 = fig3.add_subplot(122)
    ax2.set(title='Magnitude of Edges after non max. suppression')
    ax2.imshow(suppressed_image, cmap=plt.cm.gray)
    plt.show()
    
interact(non_max_suppression)
```

A low and high threshold can be used to refine the edges found in the previous step. 
After processing the image using the two thresholds, a hysteresis threshold using a 8 connected neighborhood can optionally be used to process the image further to get rid of weak edges.
Plots displayed:
1. Refined image using low and high threshold (double threshold)
    - (Optional) Refined image after undergoing hysteresis thresholding
3. Binarized Output of the previous image

```python
def hysteresis_threshold(hyst_image):
    width_of_blob = height_of_blob = 3
    # the problem with this method being that a narrow strip to the right hand side wont be covered during edge
    # detection
    for i in range(0, len(hyst_image), width_of_blob):
        for j in range(0, len(hyst_image[i]), height_of_blob):
            if 1 not in (hyst_image[i-1][j-1], hyst_image[i][j-1], hyst_image[i+1][j-1],
                     hyst_image[i-1][j], hyst_image[i+1][j],
                     hyst_image[i-1][j+1], hyst_image[i][j+1], hyst_image[i+1][j+1]):
                hyst_image[i-1][j-1], hyst_image[i][j-1], hyst_image[i+1][j-1] = 0, 0, 0
                hyst_image[i-1][j], hyst_image[i][j], hyst_image[i+1][j] = 0, 0, 0
                hyst_image[i-1][j+1], hyst_image[i][j+1], hyst_image[i+1][j+1] = 0, 0, 0
    return hyst_image
            
    
def double_threshold(high_threshold, low_threshold, bin_threshold, use_hyst):
    thresh_lim_image = suppressed_image.copy()
    for i in range(len(thresh_lim_image)):
        for j in range(len(thresh_lim_image[i])):
            if thresh_lim_image[i][j] > high_threshold:
                thresh_lim_image[i][j] = 1
            elif thresh_lim_image[i][j] < low_threshold:
                thresh_lim_image[i][j] = 0
    
    if (use_hyst):
        hyst_image = hysteresis_threshold(thresh_lim_image.copy())
        bin_image = hyst_image.copy()
    else:
        bin_image = thresh_lim_image.copy()

    for i in range(len(bin_image)):
        for j in range(len(bin_image[i])):
            if bin_image[i][j] > bin_threshold:
                bin_image[i][j] = 1
            else:
                bin_image[i][j] = 0
                
    fig4 = plt.figure()
    fig4.set_size_inches(12, 6)
    ax1 = fig4.add_subplot(131)
    ax1.set(title='Image with high/low threshold')
    ax1.axis('off')
    ax1.imshow(thresh_lim_image, cmap=plt.cm.gray)
    if (use_hyst):
        ax2 = fig4.add_subplot(132)
        ax2.set(title='After applying hysteresis')
        ax2.imshow(hyst_image, cmap=plt.cm.gray)
        ax2.axis('off')
    ax3 = fig4.add_subplot(133)
    ax3.set(title='Binarized output of image on the left')
    ax3.axis('off')
    ax3.imshow(bin_image, cmap=plt.cm.gray)
    
interact(double_threshold, high_threshold=FloatSlider(min=0.5, max=0.99, step=0.01, value=0.8, continuous_update=False),
         low_threshold=FloatSlider(min=0.01, max=0.49, step=0.01, value=0.1, continuous_update=False),
         bin_threshold=FloatSlider(min=0.01, max=0.99, step=0.01, value=0.2, continuous_update=False),
         use_hyst=Checkbox(value=False,
                           description='Use Hysteresis Thresholding',
                           disabled=False))
```
